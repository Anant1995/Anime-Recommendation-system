```python
import re
def year_extract(v):
    comp=re.compile(r"\d{4}")
    match=comp.findall(v)
    if match:
        if len(v)>1:
            return (re.findall(r"\d{4}",v))[0]
        else:
            return re.findall(r"\d{4}",v)
```

# Feature Engineering

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
        
```

```python
anime.tags=anime.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(anime.tags).toarray()
```

```python
import numpy as np
vec=np.array(cv.fit_transform(anime.tags).toarray())
```

# Using Cosine Similarity to get nearest vector

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vec)
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

# Making Recommendation Function

```python
def recommend(anime_rec):
    anime_rec_index=anime[anime.title==anime_rec].index[0]
    distance=similarity[anime_rec_index]
    anime_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in anime_list:
        print(anime.iloc[i[0]].title)
```

```python
recommend('Death Note')
```
